### Training the model

```python
df_cancer
```

```python
X = df_cancer.drop(['target'], axis = 1) # axis=1 means column wise
X.head()
```

```python
y = df_cancer['target']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
from sklearn.svm import SVC # svm: Support Vector Machine
```

```python
# Classification report: Gives the summary of our performance
# Confusion matrix: To evalte the model
```

```python
# Type-2 Error is the most dangerous, think about Cancer prediction
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

### Model Evaluation

```python
# Key objective of machine learning: To generalise the data; can be used for any dataset in real world senarios
# Overfitted model: Specially designed for a training dataset
# Generalised model works better
```

```python
y_predict = svc_model.predict(X_test)
y_predict
```

```python
cm = confusion_matrix(y_test,y_predict)
```

```python
plt.figure(figsize = (4,3))
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,y_predict))
```

### Improving the Model

```python
# Two ways: Data normalization and tune SVM classifier parameters according to data
```

#### Data Normalization

```python
min_train = X_train.min()
range_train = (X_train - min_train).max()

X_train_scaled = (X_train-min_train)/range_train
```

```python
plt.figure(figsize = (4,3))
sns.scatterplot(x=X_train['mean area'], y=X_train['mean smoothness'], hue = y_train)
```

```python
plt.figure(figsize = (4,3))
sns.scatterplot(x=X_train_scaled['mean area'], y=X_train_scaled['mean smoothness'], hue = y_train)
```

```python
min_test = X_test.min()
range_test = (X_test - min_test).max()

X_test_scaled = (X_test-min_test)/range_test
```

```python
svc_model.fit(X_train_scaled, y_train) # Fitting --> Predictions --> Confusion matrix --> Heat map
```

```python
y_predict = svc_model.predict(X_test_scaled)
y_predict
```

```python
cm = confusion_matrix(y_test,y_predict)
```

```python
plt.figure(figsize = (4,3))
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,y_predict))
```

#### Optimization of SVM Parameters

```python
# How can we penalise the model if it misclassifies the value?
# 1. C Parameter: Controls trade-off b/w classifying training points correctly and having a smooth decision boundary
# Small c: Penalty of misclassification is low / Soft margin
# Large C (Strict model): Penalty of misclassification is high / Hard margin, Potentially overfit
```

```python
# 2. Gamma Parameter: Controls the influence of training set reaches
# Large Gamma: Focus on the points close to hyperplane, close reach, closer points have much higher weight, model will be overfitted
# Small Gamma: Far reach, more generalised solution
```

```python
# How to optimise the paramters C and Gamma? 
# Use GridSearch technique present in sklearn
```

```python
from sklearn.model_selection import GridSearchCV # GridSearchCV searches within our grid
```

```python
param_grid = {'C':[0.1,1,10,100], 'gamma':[0.1,1,0.01,0.001], 'kernel':['rbf']} # rbf: Radial basis function
# Param_grid: param means parameter
```

```python
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=4)
# SVC means support vector classifier
# Verbose: How many values that we want it to display while searching for our grid.
```

```python
grid.fit(X_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
grid_predictions = grid.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test,grid_predictions)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test,y_predict))
```
